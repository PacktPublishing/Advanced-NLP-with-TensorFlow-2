```python
import matplotlib.pyplot as plt

from PIL import Image
from scipy.signal import convolve2d
import numpy as np
```

```python
tulip = Image.open("chap7-tulip.jpg") 

#convert to gray scale image
tulip_grey = tulip.convert('L')
tulip_ar = np.array(tulip_grey)

# show the image
plt.imshow(tulip_grey)
tulip_ar.shape
```

```python
# Sobel Filter
kernel_1 = np.array([[1, 0, -1],
                   [2, 0, -2],
                   [1, 0, -1]])   # Vertical edge detection kernel / filter
kernel_2 = np.array([[1, 2, 1],
                   [0, 0, 0],
                   [-1, -2, -1]])   # Horizontal edge detection kernel / filter
kernel_1.shape, kernel_2.shape
```

```python
out1 = convolve2d(tulip_ar, kernel_1)  # vertical filter
out2 = convolve2d(tulip_ar, kernel_2)  # horizontal filter
```

```python
# Create a composite image from the two edge detectors
out3 = np.sqrt(out1**2 + out2**2)
```

```python
cmap = 'viridis'
fig=plt.figure(figsize=(9, 9))
fig.add_subplot(1, 4, 1)
plt.imshow(tulip_grey, cmap=cmap)
fig.add_subplot(1, 4, 2)
plt.imshow(out1, cmap=cmap)
fig.add_subplot(1, 4, 3)
plt.imshow(out2, cmap=cmap)
fig.add_subplot(1, 4, 4)
plt.imshow(out3, cmap=cmap)
#plt.subplots_adjust(hspace=0.5)
plt.show()
```
